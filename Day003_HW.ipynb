### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
# 讀檔案
fh = open("./data/64_72hr_CH.xml", "r")
xml = fh.read()
fh.close()
```

```python
#print(xml)
import xmltodict
d = dict(xmltodict.parse(xml))
chapter=d['cwbopendata']['dataset']['locations']['location']
for chapters in chapter:
    print(chapters['locationName'])
```

```python
chapters['weatherElement'][0]['time'][0]['dataTime']
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for chapters in chapter:
    print(chapters['locationName'])
    print(chapters['weatherElement'][0]['time'][0]['dataTime'],',',chapters['weatherElement'][0]['time'][0]['elementValue']['value'],'度')
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

for i in chapters['weatherElement'][0]['time']:
    print(i['dataTime'],',',i['elementValue']['value'],'度')
```
